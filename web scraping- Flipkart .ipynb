```python
product=soup.find_all("div",class_ ="_75nlfW")
```

```python
Laptop=[]
```

```python
for i in product:
    title=i.find( "div" ,class_="KzDlHZ").get_text(strip=True)
    price=i.find("div", class_="Nx9bqj _4b5DiR").get_text(strip=True)
    features=i.find("ul", class_="G4BRas").get_text(strip=True)
    rating=i.find("div",class_="XQDdHH")
    if rating:
        ratings=rating.get_text(strip=True)

        Laptop.append({
     "Name":title,
      "Price": price,
    "Ratings" : ratings,

        })
        
```
